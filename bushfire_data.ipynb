```python
# import packages 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# read data as csv
file_bushfire = '../DISNEY-DONUT-PROJECT/Resources/bushfire_data.csv'
bushfire_df = pd.read_csv(file_bushfire)
```

```python
bushfire_df.head()
```

```python
# check data that has number is a 'float', i.e. every variable except 'year', 'description', 'state' and 'series ID' should be a float
bushfire_df.dtypes
```

```python
# check timeseries 
bushfire_df.groupby('state').count()['year'].sort_values()
```

```python
# proxy for the severity of bushfire is 'area burned ha'
# check na's in these two variables 

bushfire_df['area burned ha'].isnull().sum()
```

```python
yearly_fires = bushfire_df.groupby('year').sum()
yearly_fires
```

```python
# create dataframe that does not have any nulls for 'approx value loss'
value_loss_df = bushfire_df.dropna(subset =  ['area burned ha', 'approx value loss'])
value_loss_df.isnull().sum()
```

```python
sum_loss = value_loss_df.groupby('year').sum()
sum_loss
```
